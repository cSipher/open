```python
correlation_threshold = 0.85
correlation_matrix = df.corr()

high_corr = correlation_matrix[correlation_matrix.abs() > correlation_threshold]

plt.figure(figsize=(20, 20))
sns.heatmap(high_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Features with High Correlation")
plt.show()
```

```python
def correlation(dataset, threshold):
    cols = set()
    corr_matrix = dataset.corr()


    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                cols.add(colname)

    return list(cols)


corr_features = correlation(df, correlation_threshold)

print("Columns with high correlation (above threshold):", sorted(corr_features))
print("Number of columns with high correlation:", len(corr_features))
```

```python
df.drop(columns=corr_features,axis=1,inplace=True)
```

3. Data Preparation for Modeling

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
scaler = StandardScaler()
x_sc = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
def evaluate_model(model, y_test, y_pred):

    print(f"\n{model} Results:")
    print("Confusion Matrix:")
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{model} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")
```

4. Model Training and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
evaluate_model("Logistic Regression", y_test, lr_pred)
```

```python
# Random Forest
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
evaluate_model("Random Forest", y_test, rf_pred)
```

```python
svc = SVC()
svc.fit(x_train, y_train)
svc_pred = svc.predict(x_test)
evaluate_model("Support Vector Machine", y_test, svc_pred)
```

```python
gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
gb_pred = gb.predict(x_test)
evaluate_model("Gradient Boosting", y_test, gb_pred)
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_test)
evaluate_model("XGBoost", y_test, xgb_pred)
```

```python
# LightGBM
lgbm = LGBMClassifier()
lgbm.fit(x_train, y_train)
lgbm_pred = lgbm.predict(x_test)
evaluate_model("LightGBM", y_test, lgbm_pred)
```

```python

# Store model names and their predictions in a dictionary
model_predictions = {
    "Logistic Regression": lr_pred,
    "Random Forest": rf_pred,
    "Support Vector Machine": svc_pred,
    "Gradient Boosting": gb_pred,
    "XGBoost": xgb_pred,
    "LightGBM": lgbm_pred,
}

# Calculate accuracies and store in a dictionary
accuracies_dict = {}
for model_name, predictions in model_predictions.items():
    accuracy = accuracy_score(y_test, predictions)
    accuracies_dict[model_name] = accuracy

# Create a Pandas DataFrame from the dictionary
df_accuracies = pd.DataFrame(accuracies_dict, index=['Accuracy']).T.sort_values(by=['Accuracy'], ascending=False)
display(df_accuracies)
```

ROC Curve Analysis:

```python
from sklearn.metrics import roc_curve, auc
```

```python


# Create a dictionary of models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(probability=True),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "LightGBM": LGBMClassifier()
}

# Train and store predictions for each model
model_predictions = {}
for model_name, model in models.items():
    model.fit(x_train, y_train)
    model_predictions[model_name] = model.predict(x_test)

# Create subplots for each model
num_models = len(models)
num_cols = 3  # Number of columns in the subplot grid
num_rows = (num_models + num_cols - 1) // num_cols  # Calculate number of rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))  # Adjust figure size
fig.suptitle('ROC Curves for All Models', fontsize=16)  # Add overall title

# Plot ROC curves in each subplot
for i, (model_name, model) in enumerate(models.items()):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]  # Get the current subplot

    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(x_test)[:, 1]
    else:
        probs = model.decision_function(x_test)
    fpr, tpr, thresholds = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)

    ax.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], 'k--', label='Random Guess')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'{model_name} ROC Curve')
    ax.legend(loc='lower right')

# Hide any empty subplots
for i in range(num_models, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    axes[row, col].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to prevent overlapping
plt.show()
```
